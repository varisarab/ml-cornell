## Perform K-means on the data set to cluster the patients into 2 clusters, compare the patient cluster assignment with their ground truth labels (malignant vs. benign) and calculate the clustering accuracy.

```python
# Import pandas function

import pandas as pd
```

```python
# Import data in text file format

data_bc = pd.read_csv('wdbc.data.txt', header=None)
```

```python
data_bc
```

```python
x = data_bc.iloc[:, 2:33]
y = data_bc.iloc[:, 1]
```

```python
y
```

```python
# Apply K-Means clustering

from sklearn.cluster import KMeans
import numpy as np
estimator = KMeans(n_clusters=2, init='k-means++', n_init=10, random_state=0).fit_predict(x)
```

```python
estimator
```

```python
# Replace letters with numeric, and convert type to integers

import numpy as np
a1 = y
a1= a1.replace(["M", "B"],[1, 0])
a1.astype(int)
```

```python
# Import accuracy score functions to calculate accuracy score

from sklearn.metrics import accuracy_score
from sklearn.metrics import adjusted_rand_score
```

```python
# calculate accuracy score

accuracy_score(a1, estimator)
```

```python
# switching the classes to find which yields higher accuracy

a2 = y

a2= a2.replace(["B", "M"],[1, 0])
a2.astype(int)
```

```python
# Alternate accuracy score

accuracy_score(a2, estimator)
```

## Perform hierarchical agglomerative clustering on the data set to cluster the patients into 2 clusters using single, average and complete linkage, compare the patient cluster assignment with their ground truth labels (malignant vs. benign) and calculate the clustering accuracy for each method.

```python
# To visualize hierarchical agglomerative clustering, apply dendrogram to show different distances between different clusters
# Apply hierarchical agglomerative clustering function

import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
dendrogram(linkage(x, method='single'), leaf_rotation=90)
plt.title('Hierarchical Clustering(Nearest Point)')
plt.show()
```

```python
# Apply complete linkage for hierarchical agglomerative clustering

dendrogram(linkage(x, method='complete'), leaf_rotation=90)
plt.title('Hierarchical Clustering(Farthest Point)')
plt.show()
```

```python
# Apply average linkage for hierarchical agglomerative clustering

dendrogram(linkage(x, method='average'), leaf_rotation=90)
plt.title('Hierarchical Clustering(UPGMA)')
plt.show()
```

```python
# Use AgglomerativeClustering function to see similarity of group objects in clusters
# Apply 'complete' linkage that uses maximum distances between all observations in 2 sets

from sklearn.cluster import AgglomerativeClustering
model1 = AgglomerativeClustering(n_clusters=2, linkage='complete').fit_predict(x)
```

```python
# Calculate accuracy score for complete linkage in agglomerative clustering

accuracy_score(a1, model1)
```

```python
# Apply 'ward' linkage for variance of merged clusters

model2 = AgglomerativeClustering(n_clusters=2, linkage='ward').fit_predict(x)
```

```python
# Calculate accuracy score for ward linkage in agglomerative clustering

accuracy_score(a1, model2)
```

```python
# Apply 'average' as linkage to find average distance between observations from pairs of clusters 

model3 = AgglomerativeClustering(n_clusters=2, linkage='average').fit_predict(x)
```

```python
# Calculate accuracy score for the average linkage in agglomerative clustering

accuracy_score(a1, model3)
```

## Perform PCA (keep all components) on the data set and plot the patients as dots in the two-dimensional space spanned by the first and second principal components. Using different colors for patients in different classes (malignant vs. benign). Based on the scatter plot, try to explain if k-means can work well on this data set after PCA.
```python
# Apply StandardScaler function

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
X_scaler = StandardScaler().fit_transform(x)
```

```python
# Keeping all components
# Apply PCA function from sklearn.decomposition

import pandas as pd
from sklearn.decomposition import PCA
pca = PCA()
principalComponents = pca.fit_transform(X_scaler)
```

```python
# Plot PCA
# Visualize clusters in PCA scatter plot

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
leg_elements = [Line2D([0], [0], marker='o', color='w', label='Benign',
                          markerfacecolor='b', markersize=10),
                Line2D([0], [0], marker='o', color='w', label='Malignant',
                          markerfacecolor='r', markersize=10)]
ax.set_title('2 component PCA', fontsize = 20)
classoftumor = [1, 0]
ax.scatter(principalComponents[:, 0], principalComponents[:, 1], c = ("blue", "red"))
ax.legend(handles=leg_elements, loc='right')
ax.grid()
```

#### The PCA plot suggests that K-means clustering is not ideal on this dataset. There is no obvious, distinct separation among the features like how PCA scatter plot should appear. Since PCA is supposed to display linear uncorrelation, and K-means clustering is intended to separate the dataset into smaller groups of clusters containing features with high similarity, while keeping clusters containing dissimilar features as far apart as possible. Each cluster contains centroid at the center. When higher quality of clusters are available, best k value can be achieved. To improve the algorithm performance, first locating the initial centroids is important, and then  continuously select data points closest to centroids over many interations, adjusting the average position of the data points. This process is repeated until equilibrium was reached, and the centroids no longer change in location. K-means, as measured by Euclidean distances, is intended to minimize the sum of squared error, the clustering with lowest data points scatter in relation to their centroids. However, if assumptions about the data are not met, the Euclidean distance might not be reasonabe. We can increase K, the number of clusters, but better clustering with smaller K could produce lower sum of squared error than poor clustering with higher K as well.
